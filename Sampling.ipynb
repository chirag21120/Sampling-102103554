```python
df = df.drop(['Time'],axis=1)
```

```python
df = df.drop_duplicates()
```

```python
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
from imblearn.over_sampling import SMOTE
X_s,y_s = SMOTE().fit_resample(X,y)
```

```python
y_s.value_counts()
```

```python
data_s = pd.concat([pd.DataFrame(X_s, columns=X.columns), pd.DataFrame(y_s, columns=['Class'])], axis=1)
```

```python
data_s.head()
```

```python
X_smote=data_s.drop("Class",axis=1)
y_smote=data_s["Class"]
```

```python
Z= 1.96
p= 0.5
E= 0.05
```

```python
# Simple Random Sampling
n= ((Z**2) * p * (1 - p)) / (E**2)
```

```python
n=int(np.ceil(n))
n
```

```python
from sklearn.utils import resample
sample1 = resample(data_s, replace=False, n_samples=n)
print(sample1['Class'].value_counts())
```

```python
# Systematic sampling
N = len(data_s)
k = N // n
if N % n != 0:
  interval = k + 1
else:
  interval= k
start = np.random.randint(interval)
indexes = range(start, N, interval)
sample2 = data_s.iloc[indexes]
print(sample2['Class'].value_counts())
```

```python
#Stratified Sampling
n=392
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
stratified_sampler = StratifiedShuffleSplit(n_splits=1, test_size=n)
for sample_indices, _ in stratified_sampler.split(X_s, y_s):
    x_sample = X_s.loc[sample_indices]
    y_sample = y_s.loc[sample_indices]
sample3 = pd.concat([pd.DataFrame(x_sample, columns=X.columns), pd.DataFrame(y_sample, columns=['Class'])], axis=1)
sample3['Class'].value_counts()
```

```python
#cluster sampling
import random
num_clusters = 5

kmeans = KMeans(n_clusters=num_clusters, n_init='auto', random_state=42)

clusters = kmeans.fit_predict(data_s)
clusters = pd.Series(clusters)

selected_clusters = random.sample(range(num_clusters), 3)
sample4 = data_s.loc[clusters.isin(selected_clusters)]
sample4
```

```python
# Bootstrap Sampling
n_bootstrap = 100
BootstrapSample = data_s.copy()
sample5 = []
for _ in range(n_bootstrap):
    resampled_data = BootstrapSample.sample(n=len(data_s), replace=True, random_state=42)
    sample5.append(resampled_data)
bootstrap_means = [resampled_data.mean(axis=0) for resampled_data in sample5]
mean_estimate = np.mean(bootstrap_means, axis=0)  # Overall mean estimate
confidence_intervals = np.percentile(bootstrap_means, [2.5, 97.5], axis=0)  # 95% CIs
sample5
```
